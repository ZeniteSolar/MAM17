```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
PWM_D_MAX = int(640)
PWM_D_MAX_THRESHHOLD = int(634)
PWM_D_LIN_MULT = int(5)
PWM_D_LIN_DIV = int(1)


def pwm_compute(x: int):
    #  y = (a * x) >> b
    y = PWM_D_LIN_MULT * x / 2**PWM_D_LIN_DIV
    return y

x = np.arange(255, dtype=int)
y = pwm_compute(x)
plt.plot(x, y); plt.show()
```

```python
# 3000 -> 16s
# 0    -> 2s
# x from 0 to 255

pwm_min = 0
pwm_max = PWM_D_MAX

t_max = int(16)
t_min = int(2)
cycles_max = int(3000)
cycles_min = int(t_min * cycles_max / t_max)

y_min = cycles_min
y_max = cycles_max
x_min = pwm_min
x_max = pwm_max

# cenario 1: quando x == x_min -> y = y_min
# y_min = a * x_min + b
# cenário 2: quando x == x_max -> y = y_max
# y_max = a * x_max + b
b = int(y_min)
a = int((y_max - y_min) / x_max)

y = a*x +b

plt.plot(x, y); plt.show()

print(f"y = {a} x + {b}")
print(f"a = {a}")
print(f"b = {b}")
print(f"test y_max: ({a*x_max + b} ~= {y_max}) ?")
print(f"test y_min: ({a*x_min + b} ~= {y_min}) ?")
```
